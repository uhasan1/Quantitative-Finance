```python
import pandas as pd
```

```python
gebiz = pd.read_csv("data/gebiz/government-procurement-via-gebiz.csv")
```

```python
len(gebiz)
```

```python
gebiz.tail()
```

### 4. Display descriptive statistics for the dataset

```python
gebiz.describe(include = 'all')
```

### 5. List the top ten agencies based on number of tenders released

```python
gebiz.groupby('agency').count().sort_values(by = 'tender_no.', ascending = False)['tender_no.'].head(10)
```

### 6. List the top ten suppliers by total awarded amount

```python
gebiz.groupby('supplier_name').sum().sort_values(by = 'awarded_amt', ascending = False).head(10)
```

### 7. What percentage of the tenders were not awarded to any suppliers?

```python
len(gebiz[gebiz['supplier_name'].isnull()])/len(gebiz)
```

### 8. Create new columns for the year and month of the tender award

```python
gebiz['year'] = gebiz['award_date'].str.split('-').str[0]
gebiz['month'] = gebiz['award_date'].str.split('-').str[1]
```

### 9. List the top five months in 2016 which had the most number of tenders awarded to suppliers

```python
gebiz[gebiz['year'] == '2016'].groupby('month').count().sort_values('tender_no.', ascending = False)['tender_no.'].head()
```

### 10. List the top ten agency-supplier pairs by number of tenders awarded

```python
gebiz.groupby(['agency','supplier_name']).count().sort_values('tender_no.', ascending = False)['tender_no.'].head(10)
```
