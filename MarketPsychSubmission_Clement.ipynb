```python
## Import Python Libraries ## 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Load sentiment and prices data as sdf and pdf respectively ##
sdf = pd.read_csv('US_sentiment.csv', parse_dates=['Date'])
pdf = pd.read_csv('US_prices.csv', parse_dates=['Date'])
display(sdf)
display(pdf)
```

```python
## Compute the 30-day Moving Average news sentiment ##
sdf['ma30'] = sdf['sentiment'].rolling(30).mean()
```

```python
## Merge the 2 datasets, i.e. sdf and pdf, as pdf_new to remove non-trading days in sdf ##
pdf_new = pd.merge(pdf, sdf, how = 'left', on = ['Asset', 'Date']) # Optional parameter: Asset

## Set pdf_new index as Date ##
pdf_new.index = pdf_new['Date']

## Remove redundant column data in pdf_new ##
del pdf_new['Date'], pdf_new['Volume'], pdf_new['windowTimestamp'], pdf_new['dataType'], pdf_new['systemVersion']

## Display first 5 rows of pdf_new ##
pdf_new.head()
```

```python
## Plot the time series of asset price and 30-day MA sentiment - optional ##
plt.figure(figsize=(20,8))

ax1 = plt.subplot(2,1,1)
pdf_new['Close'].plot(label='Price', color='r')
plt.legend()

ax2 = plt.subplot(2,1,2,sharex=ax1)
pdf_new['ma30'].plot(label='MA30')
plt.legend()

plt.show()
```

```python
""" 
Assumptions: 
(1) Sentiment data of day is available at the close
(2) Prepare position using current trading day's information (i.e. Close, sentiment and ma30) and 
execute order depending on next trading day's information (i.e. Low and High) 
(3) Negligible transaction costs
"""

## Initialize Position in pdf_new ##
pdf_new['Position'] = ''

## Determine Long/Short positions: If sentiment > ma30, Position = Short. Otherwise, Position = Long ##
pdf_new.loc[(pdf_new['sentiment'] > pdf_new['ma30']) & (pdf_new['ma30'].notna()), ['Position']] = 'Short'
pdf_new.loc[(pdf_new['sentiment'] < pdf_new['ma30']) & (pdf_new['ma30'].notna()), ['Position']] = 'Long'

## Initialize Order_Execution, High_Tmr and Low_Tmr in pdf_new ##
pdf_new['Order_Execution'] = ''
lst = pdf_new['High'][1:].tolist()
lst.append(np.nan)
pdf_new['High_Tmr'] = lst
lst = pdf_new['Low'][1:].tolist()
lst.append(np.nan)
pdf_new['Low_Tmr'] = lst

"""
Determine Order_Execution: 
If Close is between Low_Tmr and High_Tmr and Position = Long, Order_Execution = True (i.e. Execute at Close)
If Close is between Low_Tmr and High_Tmr and Position = Short, Order_Execution = True (i.e. Execute at Close)
Otherwise, Order_Execution = False (i.e. Do not execute order)
"""

pdf_new.loc[(pdf_new['Close'] >= pdf_new['Low_Tmr']) & (pdf_new['Close'] <= pdf_new['High_Tmr']) & (pdf_new['Position'] == 'Long'), ['Order_Execution']] = True
pdf_new.loc[(pdf_new['Close'] >= pdf_new['Low_Tmr']) & (pdf_new['Close'] <= pdf_new['High_Tmr']) & (pdf_new['Position'] == 'Short'), ['Order_Execution']] = True
pdf_new.loc[pdf_new['Order_Execution'] == '', ['Order_Execution']] = False

## Display last 10 rows of pdf_new ##
pdf_new.tail(10)
```

```python
"""
Assumptions made: 
(4) Initial capital is $1 - Use part of it for investment and save remainder as reserve so as to allow draw down in the event of losses)
(5) Daily rebalancing at the close
"""

## Set initial capital, invest_ratio (i.e. ratio/amount to invest) and reserve  ##
initial_cap = float(1.0)
invest_ratio = float(0.839)
reserve = initial_cap * (1 - invest_ratio)

## Initialize Close_Tmr and PnL in pdf_new ##
lst = pdf_new['Close'][1:].tolist()
lst.append(np.nan)
pdf_new['Close_Tmr'] = lst
pdf_new['PnL'] = float(0.0)

"""
Create Trading Strategy:
If Position = Long and Order_Execution = True, PnL = ratio * ((Close for next trading day / Close at current trading day) - initial capital)
If Position = Short and Order_Execution = True, PnL = ratio* (initial capital - (Close for next trading day / Close at current trading day))

"""

pdf_new.loc[(pdf_new['Position'] == 'Long') & pdf_new['Order_Execution'] == True, ['PnL']] = pdf_new.apply(lambda x: invest_ratio * ((x['Close_Tmr'] / x['Close']) - initial_cap), axis=1)
pdf_new.loc[(pdf_new['Position'] == 'Short') & pdf_new['Order_Execution'] == True, ['PnL']] = pdf_new.apply(lambda x: invest_ratio * (initial_cap - (x['Close_Tmr'] / x['Close'])), axis=1)

## Compute cumulative reserve and returns ##
pdf_new['Reserve'] = pdf_new['PnL'].cumsum() + reserve
pdf_new['Returns'] = pdf_new['PnL'].cumsum() + reserve + initial_cap * invest_ratio

## Send an alert if reserve is depleted ##
if (pdf_new['Reserve'] <= float(0.0)).any():
    print('Please revise investment ratio!')
```

```python
## Compute total profits earned from trading strategy ##
print('Total profits earned: $', pdf_new['PnL'].sum())
```

```python
## Plot the time series of Returns ##
plt.figure(figsize=(20,8))
pdf_new['Returns'].plot(color = 'g')
```

```python
## Plot the time series of asset price, 30-day MA sentiment and Returns - optional ##
plt.figure(figsize=(20,8))

ax1 = plt.subplot(3,1,1)
pdf_new['Close'].plot(label='Price', color='r')
plt.legend()

ax2 = plt.subplot(3,1,2,sharex=ax1)
pdf_new['ma30'].plot(label='MA30')
plt.legend()

ax3 = plt.subplot(3,1,3,sharex=ax1)
pdf_new['Returns'].plot(label='Returns', color = 'g')
plt.legend()

plt.show()
```
