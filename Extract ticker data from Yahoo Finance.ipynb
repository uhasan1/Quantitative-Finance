```python
## Import Python libraries ## 
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import pandas_datareader.data as web
```

```python
## Indicate timeframe of ticker data ##
# For this example, we are using 1 Jan 2016 to today
start = dt.datetime(2016, 1, 1)
end = dt.datetime.now()
```

```python
## Optional - Display time-series visualizations of adjusted closing prices of stock ##

colors = ['r', 'b', 'g', 'c', 'm', 'y']

def ts_visuals(list_of_stocks, data, sequence):
    
    ## Set dimensions of time-series visualizations ##
    plt.figure(figsize=(20,8))
    
    ## Set plotting parameters ## 
    ax = plt.subplot(len(list_of_stocks), 1, sequence + 1)
    data['Adj Close'].plot(label = stock, color = colors[sequence])
    plt.legend()
```

```python
## Export sample ticker data from Yahoo finance as csv ##
list_of_stocks = ['AAPL', 'MSFT', 'GOOG']
sequence = 0

for stock in list_of_stocks:
    
    ## Export ticker data ##
    data = web.DataReader(stock, 'yahoo', start, end)
    data.to_csv(stock + ".csv")
    
    ## Optional - Display time-series visualizations ##
    ts_visuals(list_of_stocks, data, sequence)
    sequence += 1
```
